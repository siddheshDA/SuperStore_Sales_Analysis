```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_column',None)
```

```python
data = pd.read_csv("SuperStoreOrders.csv")
d = pd.DataFrame(data)
df = d.copy()
df.head()
```

```python
df.info()
```

```python
# checking nulls:
df.isnull().sum()
```

```python
df.columns
```

```python
cat_col = ['order_id', 'ship_mode', 'customer_name',
       'segment', 'state', 'country', 'market', 'region', 'product_id',
       'category', 'sub_category', 'product_name', 'order_priority', 'year']
for col in cat_col:
    df[col] = df[col].astype('category')
```

```python
# Checking is there any miss typed value?

count_col = ['ship_mode','segment', 'state', 'country', 'market', 'region',
       'category', 'sub_category', 'product_name', 'order_priority', 'year']
for col in count_col:
    plt.figure(figsize = (10,4))
    sns.countplot(data = df,x = col)
    plt.show()
```

```python
df.tail(2)
```

```python
df.info()
```

```python
# as date column include mixed date format 
date_col = ["order_date","ship_date"]
for col in date_col:
    df[col] = pd.to_datetime(df[col], errors="coerce", format="mixed")
```

```python
# changing data type of sales column:
# as there some values in sales column contains str ",":
df["sales"] = df["sales"].str.replace(",","",regex = True)
df["sales"] = df["sales"].astype('float')
```

```python
df.describe()
```

## As we can see in profit column there are values in -ve

```python
sns.violinplot(data = df,x = "profit")
```

### I am going to add a new column named as profit status (are we in loss or in profit per product)

```python
df["profit_status"] = df["profit"].apply(lambda x: "in profit" if x > 0 else ("neutral" if x == 0 else "in loss"))
df["profit_status"] = df["profit_status"].astype('category')
```

```python
df["shipping_diff (days)"] = df["ship_date"] - df["order_date"]
df["shipping_diff (days)"]  = df["shipping_diff"].dt.days
```
